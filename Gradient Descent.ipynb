## Gradient Descent

```python
* How your Linear Regression model finds out the Best Fit Line ?

Ans :- Gradient Descent
    >>> It is an optimization algorithm.
    >>> It is used to find optimal values of cost function(Evaluation Matric)
    >>> MSE is the default cost function in linear regression.
    >>> Best Fit Line:
        >> pass/goes through maximum number of datapoints
        >> It cant pass through all the datapoints
        >> It minimizes the distance between the datapoints and the line
        >> It will always pass through the point where x_mean and y_mean meets.
        >> Least sum of squares(OLS >> Ordinary Least Squares)
```

```python
>> Error/Residual >> difference between actual and predicted values

>> Loss Function >> Error for a single point
>> Cost Function >> all the errors combined

>> Sum of Squared Error/Residual Sum of Squares:
    sum((ya-yp)**2)

>> Mean Squared Error : sum((ya-yp)**2)/n
    
>> The lesser the errors, more reliable the model is

>> Variance:
    Stochastic
    Deterministic
    
>> When we use squared term in Cost Function >> Quadratic Equation:
    When we plot m & c against the error, it acquires a shape of bowl, we call it as convex fuction.
    - It gives us guarantee to have absolute minima(global minima)
    - Each point on the bowl have some m and c values and SSE which carries a line
    
>> Gradient Descent : Descenting down the gradient of bowl
    
    Learning step will become smaller and smaller as it descents down and should not oscillate/converge
    This step uses Bold Driver Algo in the Backend
    
    It internally used Partial Derivatives

>> Partial Derivatives:
    For a small change in m and c in either direction, how much the error drops or increases
    
Global Minima >> Best m and c , Lowest SSE/MSE

Local Minima >> Other lower grooves other than Global Minima(Deep Learning and Non COnvex Functions)
```
